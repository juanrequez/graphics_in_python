```python
import matplotlib.pyplot as plt
import numpy as np
```

## Creación de una gráfica simple

```python
x=np.array([1,2,3,4])
```

```python
plt.plot(x)
plt.show()
```

Véase que la gráfica no tiene propiedades, ni etiquetas, ni mayor información. Es importante asignar estos valores en la gráfica.

```python
x=np.linspace(0,2*np.pi*8,1000)
```

```python
y=np.sin(x)
w=np.cos(x)
```

```python
plt.plot(x,y,'r.',x,w,'b--')
plt.legend(['sin(x)','cos(x)'])
```

```python
#Pueden usarse subdivisiones de las gráficas
plt.figure(1)
plt.subplot(2,1,1)
plt.plot(x,y,'k')
plt.legend(["sin(x)"],loc=0)
plt.grid(True)
plt.xlabel("eje x")
plt.ylabel("eje y")
plt.subplot(2,1,2)
plt.plot(x,w,'r')
plt.legend(["cos(x)"],loc=0)
plt.grid(True)
plt.xlabel("eje x")
plt.ylabel("eje y")
```

## Graficación 3D

```python
X=np.arange(-4,4,0.1)
Y=X.copy()
```

```python
Xc,Yc=np.meshgrid(X,Y)
```

```python
Zc=np.sin(Xc**2+Yc**2)/(Xc**2+Yc**2)
```

```python
fig=plt.figure()
ax=fig.add_subplot(projection="3d")
ax.plot_surface(Xc,Yc,Zc)
```
